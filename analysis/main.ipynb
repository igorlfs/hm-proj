```python
import pandas as pd
import numpy as np
```

```python
# Genetic

lei_df = pd.read_csv('../results/genetic/lei.csv')
myc_df = pd.read_csv('../results/genetic/myc.csv')
reg_df = pd.read_csv('../results/genetic/reg.csv')
sgb_df = pd.read_csv('../results/genetic/sgb.csv')

random_50_df = pd.read_csv('../results/genetic/random/50.csv')
random_100_df = pd.read_csv('../results/genetic/random/100.csv')
random_250_df = pd.read_csv('../results/genetic/random/250.csv')
random_500_df = pd.read_csv('../results/genetic/random/500.csv')

instances = ["LEI", "MYC", "REG", "SGB"]
colors = np.array([
    lei_df['num_colors'].mean(),
    myc_df['num_colors'].mean(),
    reg_df['num_colors'].mean(),
    sgb_df['num_colors'].mean()
])
time = np.array([
    lei_df['time'].mean(),
    myc_df['time'].mean(),
    reg_df['time'].mean(),
    sgb_df['time'].mean()
])

for i, instance in enumerate(instances):
    print(f"{instance} - Colors: {colors[i]} - Time: {time[i]}")

print(f"Mean - Colors: {colors.mean()} - Time: {time.mean()}")

print()

instances_rand = ["50", "100", "250", "500"]
colors_rand = np.array([
    random_50_df['num_colors'].mean(),
    random_100_df['num_colors'].mean(),
    random_250_df['num_colors'].mean(),
    random_500_df['num_colors'].mean()
])
time_rand = np.array([
    random_50_df['time'].mean(),
    random_100_df['time'].mean(),
    random_250_df['time'].mean(),
    random_500_df['time'].mean()
])

for i, instance in enumerate(instances_rand):
    print(f"{instance} - Colors: {colors_rand[i]} - Time: {time_rand[i]}")

print(f"Mean - Colors: {colors_rand.mean()} - Time: {time_rand.mean()}")
```
